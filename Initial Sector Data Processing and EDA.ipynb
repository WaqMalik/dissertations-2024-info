```python
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_investors = pd.read_csv('investors.csv', low_memory=False)
df_investments = pd.read_csv('investments.csv')
df_funding_rounds = pd.read_csv('funding_rounds.csv')
df_geo2 = pd.read_csv('investments_investors_funding_geo2.csv')
df_organisations = pd.read_csv('organizations.csv')
```

```python
df_investments_sectors = pd.merge(df_geo2, df_organisations, how='left', left_on='org_uuid', right_on='uuid')
df_investments_sectors
```

```python
df_investments_sectors.columns
```

```python
df_sectors = df_investments_sectors[['investment_uuid', 'investment_name', 'investment_type',
       'investor_uuid', 'investor_name', 'investor_type', 'announced_on',
       'is_lead_investor', 'investor_country_code', 'investor_state_code',
       'investor_region', 'investor_city', 'investor_types',
       'investor_num_of_investments', 'founded_on_x', 'closed_on_x', 'fr_uuid',
       'name_x', 'type_x', 'org_country_code', 'org_state_code', 'org_region',
       'org_city', 'fr_num_of_investors', 'org_uuid', 'org_name',
       'lead_investor_uuids', 'category_list', 'category_groups_list' ]]
df_sectors
```

```python
category_groups_split = df_investments_sectors['category_groups_list'].str.split(',')
all_category_groups = [item.strip() for sublist in category_groups_split.dropna() for item in sublist]
unique_category_groups = set(all_category_groups)
print(unique_category_groups)
```

```python
for category in unique_category_groups:
    df_sectors[category] = df_investments_sectors['category_groups_list'].apply(lambda x: 1 if pd.notna(x) and category in x.split(',') else 0)
df_sectors
```

```python
grouped_df = df_sectors.groupby('investor_uuid')[list(unique_category_groups)].sum().reset_index()
grouped_df
```

```python
grouped_df.to_csv('sector_grouped_investors.csv', index=False)
```

```python
category_columns = list(unique_category_groups)
grouped_df[category_columns] = grouped_df[category_columns].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
grouped_df.to_csv('normalized_sector_grouped_investors.csv', index=False)
```

```python
grouped_df
```

```python
#List of category columns
category_columns = grouped_df.columns.difference(['investor_uuid'])

# Create a new column 'total' by summing up all category columns
grouped_df['total'] = grouped_df[category_columns].sum(axis=1)

# Save the updated DataFrame to a new CSV file
grouped_df.to_csv('sector_grouped_investors_with_total.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('sector_grouped_investors_with_total.csv')

# List of category columns (excluding 'investor_uuid' and 'total')
category_columns = df.columns.difference(['investor_uuid', 'total'])

# Sum the total investments for each sector
sector_totals = df[category_columns].sum()

# Plot Bar chart
plt.figure(figsize=(12, 8))
sector_totals.plot(kind='bar')
plt.xlabel('Sector')
plt.ylabel('Total Investments')
plt.title('Total Investments per Sector')
plt.xticks(rotation=90)
plt.show()
```
